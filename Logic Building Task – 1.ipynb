**1. User Login Check**

```python
# Stored username and password (credentials)
valid_username = "admin"
valid_password = "1234"

# take input from user
input_username = input("Enter Username:")
input_password = input("Enter Password:")

# Check the credentials 
if valid_username == input_username and valid_password == input_password:
    print("Login Successful") # Correct
else:
    print("Invalid Credentials") # Incorrect
```

**2. Pass / Fail Analyzer**

```python
# Given marks
marks = [45, 78, 90, 33, 60]

#Total Students
total_count = len(marks)

#Initialize variables to count the pass/fail students
pass_count = 0
fail_count = 0

# Analyze results
for mark in marks:
    # Check if student passed
    if mark >= 50:
        pass_count += 1 # Increment the pass count
    # Else student failed
    else:
        fail_count += 1 # Increment the fail count

# Print the final result
print("Total Students:", len(marks))
print("Passed Students:", pass_count)
print("Failed Students:", fail_count)
```

**3. Simple Data Cleaner**

```python
# Given list
names = [" Alice ", "bob", " CHARLIE "]

# Initialize an empty list to store the cleaned name
cleaned_list = []

for name in names:
    # Remove extra spaces and convert all names to lowercase
    cleaned_name = name.strip().lower()
    # Add cleaned name to cleaned list
    cleaned_list.append(cleaned_name)

# Print cleaned list
print("Given List:", names)
print("Cleaned List:", cleaned_list)
```

**4. Message Length Analyzer**

```python
# List of messages given
messages = ["Hi", "Welcome to the platform", "OK"]

# Analyze message sizes
for message in messages:
    # Print message
    print("Message:", message)
    # Print length of message
    print("Length:", len(message))
    
    # Check and flag messages longer than 10 characters
    if len(message) > 10:
        print("Message is longer than 10 characters")
    print() # For better readability
```

**5. Error Message Detector**

```python
# Given logs
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Initialize variable to count the "ERROR" entries
error_count = 0

# Count "ERROR" entries
for log in logs:
    if log == "ERROR":
        error_count += 1

# Print total "ERROR" entries
print("Total ERROR entries:", error_count)
```
