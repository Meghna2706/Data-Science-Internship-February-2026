**Problem Statement 1:Smart Parking Lot Management System**

```python
# Function: Management system for parking lots
def smart_parking_lot(capacity, logs):
    vehicles = 0
    peak = 0

    # Calculate parked vehicles
    for log in logs:
        if log == "IN":
            vehicles += 1
        elif log == "OUT":
            if vehicles > 0:
                vehicles -= 1

        # Track peak usage
        if vehicles > peak:
            peak = vehicles

    print("\nCurrently Parked Vehicles:", vehicles)
    print("Peak parking usage:", peak)

    # Alert if parking exceeds capacity
    if vehicles > capacity:
        print("Parking Status: Exceeded Capacity")
    else:
        print("Parking Status: Available")

capacity = int(input("Parking Capacity: "))
vehicle_logs = []
vehicle_logs_number = int(input("Enter the number of vehicle logs: "))

for i in range(vehicle_logs_number):
    log = input("Enter the log (IN/OUT): ")
    vehicle_logs.append(log)

print("\nVehicle Logs:", vehicle_logs)
smart_parking_lot(capacity, vehicle_logs)
```

**Problem Statement 2:Online Food Delivery Time Estimator**

```python
# Function: Estimate online food delivery time
def delivery_time(distance, traffic, weather):
    total_time = 5 # Basic timing
    total_time += distance * 5

    # Traffic conditions
    if traffic == "Low":
        total_time += 1
    elif traffic == "Medium":
        total_time += 3
    elif traffic == "High":
        total_time += 5

    # Weather conditions
    if weather == "Clear":
        total_time += 0
    elif weather == "Rainy":
        total_time += 5
    elif weather == "Stormy":
        total_time += 10

    # Print result
    print("\nEstimated Delivery Time:", total_time, "minutes")

distance = int(input("Enter Distance: "))
traffic = input("Enter Traffic (Low/Medium/High): ")
weather = input("Enter Weather (Clear/Rainy/Stormy): ")
delivery_time(distance, traffic, weather)
```

**Problem Statement 3:Movie Theatre Seat Occupancy Analyzer**

```python
# Function: Analyzing movie theatre seat
def seat_occupancy_analyzer(total_seats, booked_seats):

    # Calculate occupancy percentage
    occupancy = (len(booked_seats) / total_seats) * 100
    print("\nTotal Seats:", total_seats)
    print("Booked Seats:", len(booked_seats))
    print("\nOccupancy:", round(occupancy), "%")

    # Determine show status
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")

    # Suggest opening additional shows
    if occupancy >= 90:
        print("Suggestion: Consider opening an additional show.")

total_seats = int(input("Enter Total Seats: "))
booked_count = int(input("Enter Number of Booked Seats: "))
booked_seats = [1] * booked_count
seat_occupancy_analyzer(total_seats, booked_seats)
```

**Problem Statement 4:Cloud Server Load Classification System**

```python
# Function: Load classifier
def classify_load(cpu_readings):
    # Calculate average CPU usage
    total = 0
    for reading in cpu_readings:
        total += reading
    average = total / len(cpu_readings)
    print("Average CPU Load:", round(average), "%")

    # Average CPU < 50% → Normal
    # 50%–80% → Warning
    # 80% → Critical
    if average < 50:
        print("Server Status: Normal")
    elif 50 <= average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")

cpu_readings = []
cpu_readings_number = int(input("Enter the number of CPU readings: "))
for i in range(cpu_readings_number):
    reading = int(input("Enter CPU reading: "))
    cpu_readings.append(reading)
print("\nCPU Readings:", cpu_readings)
classify_load(cpu_readings)
```

**Problem Statement 5:Smart Classroom Resource Usage Monitor**

```python
# Function: Resource usage monitor
def resource_usage_monitor(resource_usage):
    overused = []
    usage_limit = 8

    # Check each resource
    for resource, hours in resource_usage.items():
        if hours > usage_limit:
            overused.append(resource)

    # Check if the resources are overused or not
    if overused:
        print("\nOverused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("\nOverused Resources: None")
        print("Energy Alert: No")

resource_usage = {}
resource_usage_number = int(input("Enter the number of resources: "))

for i in range(resource_usage_number):
    resource = input("Enter resource name: ")
    hours = int(input("Enter hours used: "))
    resource_usage[resource] = hours

print("\nResource Usage:", resource_usage)
resource_usage_monitor(resource_usage)
```

**Problem Statement 6: Online Event Registration Capacity Controller**

```python
# Function: Registration capacity controller
def registration_controller(capacity, total_registrations):
    confirmed = 0
    waitlisted = 0
    # Check registrations
    for i in range(1, total_registrations + 1):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    # Print results
    print("\nConfirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)

    # Registration status
    if confirmed >= capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")

capacity = int(input("Enter Event Capacity: "))
registrations = int(input("Enter Total Registrations: "))
registration_controller(capacity, registrations)
```
