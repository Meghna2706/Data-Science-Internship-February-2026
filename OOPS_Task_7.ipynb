**Problem 1: Smart Light Controller**

```python
# Clas: Controlling a smart light that can be turned ON or OFF
class SmartLight:
    def __init__(self, light_name):
        self.name = light_name
        self.status = "OFF"

    # Display its status
    def control(self, action):
        self.status = action.upper()
        print(f"{self.name} is {self.status}")

light_name = input("Light Name: ")
action = input("Action: ")
light = SmartLight(light_name)
light.control(action)
```

**Problem 2: Employee ID Card System**

```python
# Class: Storing employee name, ID, and department
class Employee:
    def __init__(self, name, id, department):
        self.name = name
        self.id = id
        self.department = department

    # Displays ID card details
    def display_details(self):
        print("\nEmployee ID Card")
        print("Employee Name:", self.name)
        print("Employee ID:", self.id)
        print("Department:", self.department)

print("Enter Employee details")
name = input("Name: ")
id = input("ID: ")
department = input("Department: ")
employee = Employee(name, id, department)
employee.display_details()
```

**Problem 3: Mobile Contact Record**


```python
# Class: Storing contact name and phone number
class Contact:
    def __init__(self, contact_name, phone_number):
        self.contact_name = contact_name
        self.phone_number = phone_number

    # Display contact information
    def display(self):
        print("\nContact Saved")
        print("Name:", self.contact_name)
        print("Phone:", self.phone_number)

print("Enter Contact Details")
contact_name = input("Contact Name: ")
phone_number = input("Phone Number: ")
contact = Contact(contact_name, phone_number)
contact.display()
```

**Problem 4: Product Price Tag Generator**

```python
# Class: Storing product name and price
class Product:
    def __init__(self, product_name, price):
        self.product_name = product_name
        self.price = price

    # Prints a formatted price tag
    def display(self):
        print("\nPrice Tag")
        print(f"Product: {self.product_name}")
        print(f"Price: ₹{self.price}")

product_name = input("Product Name: ")
price = input("Price: ")
product = Product(product_name, price)
product.display()
```

**Problem 5: Movie Rating Display System**

```python
# Class: Storing movie name and rating
class Movie:
    def __init__(self, movie_name, rating):
        self.movie_name = movie_name
        self.rating = rating

    # Displays movie details
    def display(self):
        print("\nMovie Details")
        print("Movie:", self.movie_name)
        print("Rating:", self.rating, "/ 5")

print("Enter Movie Details")
movie_name = input("Movie Name: ")
rating = input("Rating: ")
movie = Movie(movie_name, rating)
movie.display()
```

**Problem 6: Delivery Address Manager**

```python
# Class: Storing customer name and delivery address
class Delivery:
    def __init__(self, name, address):
        self.name = name
        self.address = address

    # Prints delivery details
    def display(self):
        print("\nDelivery Details")
        print("Customer:", self.name)
        print("Address:", self.address)

print("Enter Delivery Details")
name = input("Customer Name: ")
address = input("Address: ")
delivery = Delivery(name, address)
delivery.display()
```
