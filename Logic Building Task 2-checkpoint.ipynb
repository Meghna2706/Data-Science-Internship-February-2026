**Problem 1: Unique Words in a Sentence**

```python
# Take input
sentence = input("Enter sentence: ")
print(sentence)

# Splitting sentence into words
words = sentence.split()
# Splitted sentence will be printed in the form of list
print(words)

# Converting list to set to remove duplicates
unique_words = set(words)

print() # For better rediability

# Printing both the set and its length
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

**Problem 2: Highest Salary from Employee Data**

```python
# Employee dictionary - Keys represent employee names, Values represent employee salaries
employees = {}
n = int(input("Enter number of employees:"))
print(n)
for i in range(n):
    print() # For better readability
    print("Employee", i+1, "details")
    key = input("Name:")
    print("Name:", key)
    value = int(input("Salary:"))
    print("Salary:", value)
    employees[key] = value

# Printing employee dictionary
print()
print(employees)

# Initializing the variables
highest_sal = 0
emp_name = ""

# Finding employee with highest salary
for name,salary in employees.items():
    if salary > highest_sal:
        highest_sal = salary
        emp_name = name

# Printing result
print()
print("Highest Salary:", emp_name, "-", highest_sal)
```

**Problem 3: Find Maximum and Minimum Values**

```python
# List of integers
numbers = []
n = int(input("Enter number of integers:"))
print(n)
num = 0
print("Enter the numbers")
for i in range(n):
    num = int(input())
    numbers.append(num)

# Printing the stored list
for i in numbers:
    print(i)
print()

# Initialize both maximum and minimum with the first element of the list
max_value = numbers[0]
min_value = numbers[0]

# Loop to compare each element
for num in numbers:
    if num > max_value:
        max_value = num
    if num < min_value:
        min_value = num

# Printing results
print("List:", numbers)
print("Max:", max_value)
print("Min:", min_value)
```

**Problem 4: Count Products Above a Price Threshold**

```python
# List of prices
prices = []
n = int(input("Enter the number of prices:"))
print(n)
print("Enter the prices")
price = 0
for i in range(n):
    price = int(input())
    prices.append(price)

# Printing the stored list
for i in prices:
    print(i)
print()

# Initialize counter
count = 0

# loop to compare each price with 1000
for price in prices:
    if price > 1000:
        count += 1

# Printing result
print("Products above 1000:", count)
```

**Problem 5: Calculate Attendance Percentage**

```python
# Given list
attendance = ["P", "P", "A", "P", "P"]

# Count present days
count = 0
total = len(attendance)

for status in attendance:
    if status == "P":
        count += 1

# Calculate percentage
percentage = (count / total) * 100

# Printing result
print("Attendance Percentage:", percentage)
```

**Problem 6: Remove Duplicate Phone Numbers**

```python
# List of phone numbers
phone_numbers = []
n = int(input("Enter the number of phone numbers:"))
print(n)
print("Enter the phone numbers")
phone_number = 0
for i in range(n):
    phone_number = int(input())
    phone_numbers.append(phone_number)

# Printing the stored list
for i in phone_numbers:
    print(i)
print()

# Remove duplicates using set
unique_phn_numbers = set(phone_numbers)

# Printing result
print("Unique phone numbers:", unique_phn_numbers)
```

**Problem 7: Count Character Frequency**

```python
# Given string
text = input("Enter a string:")
print(text)

# Initializing empty dictionary
char_count = {}

# Loop through each character
for ch in text:
    if ch in char_count:
        char_count[ch] += 1 # Increment the count for each character occurrence
    else:
        char_count[ch] = 1

# Print result
print(char_count)
```

**Problem 8: Convert List to Tuple**

**Problem 9: Check if a Key Exists in a Dictionary**

**Problem 10: Calculate Average Marks**

```python
# list of marks
marks = []
n = int(input("Enter number of marks:"))
print(n)
mark = 0
print("Enter the marks")
for i in range(n):
    mark = int(input())
    marks.append(mark)

# Printing the stored list
for i in marks:
    print(i)
print()

# Calculate average
average = sum(marks) / len(marks)

# Print result
print("Average Marks:", average)
```
