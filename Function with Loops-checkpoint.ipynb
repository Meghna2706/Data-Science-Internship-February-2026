# Function with Loops

**Problem 1: Mobile Recharge Validation System**

```python
# Function: Recharge validation
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    # Check whether the amount is ≥ 50
    if amount < 50:
        print("Recharge amount must be at least ₹50")
        return False
    if amount not in valid_plans:
        print("Invalid plan selected")
        print("Available plans are:", valid_plans)
        return False
    return True
    
# Retry system if invalid using while loop
while True:
    print("Valid plans: [199, 299, 399, 599]")
    # Taking recharge amount
    amount = int(input("Enter recharge amount: ₹"))
    if validate_recharge(amount):
        print("Recharge Successful")
        break
    else:
        print("Please try again.\n")
```

**Problem 2: Inventory Reorder Alert System**

```python
# Function: Inventory check
def check_inventory(stock_data):
  for product,quantity in stock_data.items():
    if quantity < 15:
      print(f"{product}:Recorder Alert(Stock = {quantity})")
    else:
      print(f"{product}:Stock OK(Stock = {quantity})")
        
# Taking input from user
inventory = {}
n = int(input("Enter number of products:"))
for i in range(n):
    product_name = input("Enter product name:")
    quantity = int(input("Enter Stock quantity:"))
    inventory[product_name] = quantity
# Printing inventory
print("Inventory:",inventory)
print("\nInventory Status:")
check_inventory(inventory)
```

**Problem 3: Student Result Processing System**

```python
# Function: Process student result
def process_result(marks):
    total = 0
    # Calculate total using loop
    for mark in marks:
        total += mark
    average = total / len(marks)
    # Decide result
    if average >= 50:
        return "Pass"
    else:
        return "Fail"

# Enter student marks
student_marks = []
n = int(input("Enter number of subjects:"))
for i in range(n):
    mark = float(input(f"Enter marks for subject {i+1}:"))
    student_marks.append(mark)
result = process_result(student_marks)

# Printing result
print()
print("Average Marks:", sum(student_marks) / len(student_marks))
print("Result:", result)
```

**Problem 4: Cab Fare Estimator with Retry Option**

```python
# Function: calculate fare
def calculate_fare(distance, peak):
    base_fare = 50
    fare = base_fare + (12 * distance)
    # Add 25% extra fare for peak hour
    if peak == "yes":
        fare = fare + (0.25 * fare)
    return fare

# Retry system
while True:
    distance = float(input("Enter distance in km:"))
    peak = input("Is it peak hour?(yes/no):").lower()
    total_fare = calculate_fare(distance, peak)
    print("Total Fare: ₹", total_fare, "\n")
    retry = input("Do you want to calculate again?(yes/no):").lower()
    if retry != "yes":
        print("\nThank you for using Cab Fare Estimator")
        break
```

**Problem 5: Employee Attendance Eligibility Checker**

```python
# Function: Check eligibility
def check_attendance(attendance_list):
    present_count = 0
    # Count present days using loop
    for day in attendance_list:
        if day == "P":
            present_count += 1
    percentage = (present_count / len(attendance_list)) * 100
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

# Taking input
attendance = []
n = int(input("Enter total working days: "))
for i in range(n):
  status = input(f"Enter attendace for day {i+1}(P/A): ")
  attendance.append(status)
result = check_attendance(attendance)
print("\nAttendance Percentage:", (attendance.count("P") / len(attendance)) * 100, "%")
print("Eligibility Status:", result)
```

**Problem 6: Password Strength Checker**

```python
# Function: Check password strength
def check_password(password):
    digit = False
    special = False
    if len(password) < 8:
        return "Weak Password(Minimum 8 characters required)"
    # Check each character
    for ch in password:
        if ch.isdigit():
            digit = True
        if ch in "@#$":
            special = True
    # Check conditions
    if digit and special:
        return "Strong Password"
    else:
        return "Weak Password(Must contain digit and special character)"

# Take input
pwd = input("Enter password: ")
result = check_password(pwd)

# Print result
print(result)
```
