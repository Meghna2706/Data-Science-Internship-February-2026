# Function-Based Problem Solving

Problem 1: Social Media – Post Engagement Analyzer

```python
# Function: Post engagement analyzer
def analyze_post_engagement(likes_list):
    total_likes = 0

    # Calculate total likes
    for like in likes_list:
        total_likes += like

    # If Total Likes ≥ 1000 → Viral Post
    # Otherwise → Normal Engagement
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    # Print result
    print("Total Likes:", total_likes)
    print("Post Status:", status)

# List of likes
likes = []
n = int(input("Enter the number of likes: "))
for i in range(n):
    like = int(input(f"Enter the number of likes for post {i+1}: "))
    likes.append(like)
print()
analyze_post_engagement(likes)
```

Problem 3: Agriculture – Rainfall Adequacy Checker

```python
# Function: Rainfall Adequacy Checker
def rainfall_checker(rainfall_list, required_level):
    total_rainfall = 0
    status = ""
    # Calculate total rainfall
    for rain in rainfall_list:
        total_rainfall += rain
    # Calculate average rainfall
    average_rainfall = total_rainfall / len(rainfall_list)

    # Average ≥ required level → Adequate Rainfall
    # Otherwise → Inadequate Rainfall
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    # Print result
    print("Average Rainfall:", round(average_rainfall))
    print("Rainfall Status:", status)


# List of seasonal rainfall per month
rainfall_list = []
n = int(input("Enter the number of months: "))
for i in range(n):
    rain = float(input(f"Enter the rainfall for month {i+1}: "))
    rainfall_list.append(rain)
print()
required_level = float(input("Enter the required rainfall level: "))
print()
rainfall_checker(rainfall_list, required_level)
```

Problem 4: Social Media – Duplicate Account Detection

```python
def detect_duplicate_accounts(users):
    # Use set removes duplicates automatically
    unique_users = set(users)
    result = ""

    # If duplicates exist → Flag accounts
    # Else → Accounts are unique
    if len(users) != len(unique_users):
        result = "Yes"
    else:
        result = "No"
    print("Duplicate Accounts Found:", result)

# List of users
users = []
n = int(input("Enter the number of users: "))
for i in range(n):
    user = input(f"Enter the username for user {i+1}: ")
    users.append(user)
print()
detect_duplicate_accounts(users)
```

Problem 6: Agriculture – Premium Crop Price Filter

```python
# Function to filter the Premium crop price
def filter_premium_crops(price_list):
    premium_crops = []

    # Price > ₹2000 → Premium Crop
    # Else → Regular Crop
    for price in price_list:
        if price > 2000:
            premium_crops.append(price)

    print("Premium Crops:", premium_crops)


# List of prices
crop_prices = []
n = int(input("Enter the number of crops: "))
for i in range(n):
    price = int(input(f"Enter the price for crop {i+1}: "))
    crop_prices.append(price)
print()
filter_premium_crops(crop_prices)
```

Problem 7: System Monitoring – Application Health

```python
# Function to check the health
def application_health_checker(error_count):

    # Errors = 0 → Healthy
    # Errors ≤ 5 → Minor Issues
    # Errors > 5 → Critical Issues
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    # Print result
    print("Error Count:", error_count)
    print("System Status:", status)

error_count = int(input("Enter the number of errors: "))
print()
application_health_checker(error_count)
```

Problem 8: Banking – Daily Transaction Limit Checker

```python
def check_transaction_limit(amount):
    daily_limit = 50000

    # Amount within limit → Approved
    # Amount exceeds limit → Rejected
    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"

    # Print result
    print("Transaction Amount:", amount)
    print("Transaction Status:", status)

amount = int(input("Enter the transaction amount: "))
print()
check_transaction_limit(amount)
```

Problem 9: E-Learning – Student Attendance Eligibility

```python
# Function to check the eligibility
def check_attendance_eligibility(attendance_list):
    total_days = len(attendance_list)
    present_days = 0

    # Attendance ≥ 75% → Eligible
    # Attendance < 75% → Not Eligible
    for day in attendance_list:
        if day == "P":
            present_days += 1

    # Calculate attendance percentage
    attendance_percentage = (present_days / total_days) * 100
    # Check eligibility
    if attendance_percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    # Print result
    print("Attendance Percentage:", attendance_percentage)
    print("Exam Eligibility:", status)


# List of attendance
attendance_list = []
n = int(input("Enter the nuPmber of days: "))
for i in range(n):
    day = input(f"Enter the attendance for day {i+1}: ")
    attendance_list.append(day)
print()
check_attendance_eligibility(attendance_list)
```

Problem 10: Smart Electricity Bill Analyzer

```python
# Function to analyze the electricity bill
def electricity_bill_analyzer(units):
    bill = 0

    # Billing calculation using slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Total bill < ₹500 → Low Usage
    # ₹500 – ₹1500 → Moderate Usage
    # ₹1500 → High Usage
    if bill < 500:
        status = "Low Usage"
    elif 500 <= bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status

units_consumed = int(input("Enter the units consumed: "))
total_bill, usage_status = electricity_bill_analyzer(units_consumed)
print()

# Print result
print("Units Consumed:", units_consumed)
print("Total Bill: ₹", total_bill)
print("Usage Level:", usage_status)
```
